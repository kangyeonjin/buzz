```python
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
import re

# PDF 파일 로드
loader = PyPDFLoader("./data/토이스토리1.pdf")
document = loader.load()

# 문서의 모든 페이지 내용 결합
all_text = "\n".join([doc.page_content for doc in document])

# 텍스트 분할기 설정(청크 크기 1000자, 중첩 0자)
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

# 텍스트 분할
texts = text_splitter.split_text(all_text)

# '버즈' 대사만 추출하기 위한 정규 표현식 패턴
buzz_pattern = re.compile(r'버즈.*?[\n.?!]')

# '버즈' 대사만 추출
# buzz_dialogues = [text for text in texts if '버즈' in text]

# 각 청크에서 '버즈' 대사만 추출
buzz_dialogues = []
for text in texts:
    # 각 청크에서 '버즈' 대사를 찾기
    buzz_dialogues.extend(re.findall(buzz_pattern, text))

# 결과 출력
for buzz in buzz_dialogues:
    print(buzz)
```

```python
import re

# 버즈 대사 목록 (단순화된 예시)
buzz_dialogues = [
    "To infinity and beyond",
    "나는 버즈 라이트이어, 그리고 난 이 부대를 관리하고 있다",
    "이건 내가 다루는 영역이야!",
    "내 마음의 소리에 귀기울여",
    ""
]

# 사용자가 질문할 수 있는 예시
user_questions = [
    "우리는 어디로 가는거야?",
    "버즈, 우주로 간다고 했지?",
    "버즈, 너는 누구야?"
]

# 대답을 찾는 함수
def respond_to_buzz(user_input):
    # 질문을 통해 버즈 대사에 맞는 대답을 찾기 (간단한 예시)
    if "우주" in user_input:
        return buzz_dialogues[0]  
    elif "누구야" in user_input:
        return buzz_dialogues[1] 
    else:
        return "그건 잘 모르겠어. 다시 말해줄래?"

# 대화 시작
for question in user_questions:
    print(f"사용자: {question}")
    response = respond_to_buzz(question)
    print(f"버즈: {response}")
```
